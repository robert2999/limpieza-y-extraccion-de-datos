```python
# leer los datos 
import pandas as pd
df=pd.read_csv(r"C:\Users\Rodenas\Downloads\medallas_.csv")
df
```

```python
# cambiar los n.a por cero
valores={"Oro":0,"Plata":0,"Bronce":0}
orden=df.fillna(valores)
orden
```

```python
orden=orden.sort_values(by="Oro", ascending=False)
orden
```

```python
filtro=orden["Oro"].idxmax()
pais=orden["Pais"][filtro]
pais
```

```python
prim=orden["Oro"]<39
ind=orden[prim]
#obtengo una serie de booleanos
indice=ind["Oro"].idxmax()
#ahora, quiero de la serie de booleanos la cantidad de medallas de oro
paise2=orden["Pais"][indice]
paise2
```

```python
prim1=orden["Oro"]<38
ind1=orden[prim1]
indice1=ind1["Oro"].idxmax()
pais1=orden["Pais"][indice1]
pais1
```

```python
resultados = {
    "Pais": [pais, paise2, pais1],
    "Oro": [orden["Oro"][filtro], orden["Oro"][indice], orden["Oro"][indice1]],
    "Plata":[orden["Plata"][filtro], orden["Plata"][indice], orden["Plata"][indice1]],
    "B"
    "Total":[orden["Total"][filtro], orden["Total"][indice], orden["Total"][indice1]]
}

df_resultado = pd.DataFrame(resultados)
df_resultado
```
